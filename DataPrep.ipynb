```python
games_and_apps = appearances.merge(games, on=['game_id'], how='left')

def player_stats(player_id, season, df):
    
    df = df[df['player_id'] == player_id]
    df = df[df['season'] == season]
    
    if (df.shape[0] == 0):
        Out = [(np.nan, season,0,0,0,0,0,0,0)]
        out_df = pd.DataFrame(data = Out, columns = ['player_id','season','goals','games',
                                                     'assists','minutes_played','goals_for','goals_against','clean_sheet'])
        return out_df
    
    else:
        
        df["goals_for"] = df.apply(lambda row: row['home_club_goals'] if row['home_club_id'] == row['player_club_id'] 
                      else row['away_club_goals'] if row['away_club_id'] == row['player_club_id'] 
                      else np.nan, axis=1)
        df["goals_against"] = df.apply(lambda row: row['away_club_goals'] if row['home_club_id'] == row['player_club_id'] 
                      else row['home_club_goals'] if row['away_club_id'] == row['player_club_id'] 
                      else np.nan, axis=1)
        df['clean_sheet'] = df.apply(lambda row: 1 if row['goals_against'] == 0
                      else 0 if row['goals_against'] > 0
                      else np.nan, axis=1)
        
        df = df.groupby(['player_id',"season"],as_index=False).agg({'goals': 'sum', 'game_id': 'nunique', 
                                                                      'assists': 'sum', 'minutes_played' : 'sum', 'goals_for' : 'sum',
                                                                      'goals_against' : 'sum', 'clean_sheet' : 'sum'})
        out_df = df.rename(columns={'game_id': 'games'})

        return out_df
    

season = 2020
for index in players.index:
    id = players.loc[index][0]
    name = players.loc[index][1]
    stats = player_stats(id, season, games_and_apps)
    players.at[index,'goals_{}'.format(season)]= stats['goals'][0]
    players.at[index,'games_{}'.format(season)]= stats['games'][0]
    players.at[index,'assists_{}'.format(season)]= stats['assists'][0]
    players.at[index,'minutes_played_{}'.format(season)]= stats['minutes_played'][0]
    players.at[index,'goals_for_{}'.format(season)]= stats['goals_for'][0]
    players.at[index,'goals_against_{}'.format(season)]= stats['goals_against'][0]
    players.at[index,'clean_sheet_{}'.format(season)]= stats['clean_sheet'][0]
```

```python
df1 = df.copy()

#  Sort the DataFrame based on player and season
df1.sort_values(by=['name', 'season'], inplace=True)

# Add a new column indicating the team the player transferred from in the previous season
df1['team_transferred_from'] = df1.groupby('player_id')['club_name'].shift(1)

# Add a new column indicating the team the player transferred from in the previous season
df1['team_transferred_from_id'] = df1.groupby('player_id')['club_id'].shift(1)

# Set 'team_transferred_from' to N/A if 'squad' and 'team_transferred_from' are the same
df1.loc[df1['club_name'] == df1['team_transferred_from'], 'team_transferred_from'] = 'N/A'

# Set 'team_transferred_from' to N/A if 'squad' and 'team_transferred_from' are the same
df1.loc[df1['club_id'] == df1['team_transferred_from_id'], 'team_transferred_from_id'] = 'N/A'

# Mark the first season for each player as not having a team transfer
df1.loc[df1.groupby('player_id').head(1).index, 'team_transferred_from'] = 'N/A'

# Mark the first season for each player as not having a team transfer
df1.loc[df1.groupby('player_id').head(1).index, 'team_transferred_from_id'] = 'N/A'

# Creating the 'team_changed' column
df1['team_changed'] = df1.groupby('player_id')['club_id'].transform(lambda x: x.ne(x.shift())).astype(int)

# Mark the first season for each player as not having a team changed
df1.loc[df1.groupby('player_id')['team_changed'].head(1).index, 'team_changed'] = 0

# Assuming you have a column 'transfer_value' representing the market value at the time of transfer
# Create a new column 'value_at_transfer' by shifting 'transfer_value' for rows where 'team_changed' is True
df1['previous_value'] = df1.groupby('player_id')['value'].shift(1)


# Fill NaN values in 'value_at_transfer' with 0 only for rows where 'team_changed' is 0
#df1.loc[df1['team_changed'] == 0, 'value_at_transfer'] = 0
df1
# Save the updated DataFrame to a CSV file
#df1.to_csv('19-22_data.csv', index=False)
```

```python
# Success Score
# Assign weights for the combination
weight_minutes = 0.5
weight_value = 0.5

# Calculate the success score with conditional calculation for the value portion
df1['success_score'] = (
    (df1['team_changed'] * 
    df1.apply(lambda row: row['value'] / row['previous_value'] if row['value'] < row['previous_value'] else 1, axis=1) * 
    weight_value +
    df1['minutes_percentage'] * weight_minutes
) * 100
)

# Fill NaN values with 0 before rounding
df1['success_score'] = df1['success_score'].fillna(0)
# Round the success_score to the nearest whole number
df1['success_score'] = df1['success_score'].round().astype(int)

# If team_changed is 0, set success_score to 0
df1.loc[df1['team_changed'] == 0, 'success_score'] = 0
```

```python
# Assuming df is your DataFrame containing all games
matches_df = pd.read_csv('input/games.csv') 
# Assuming matches_df is your DataFrame with columns ['home_team_id', 'away_team_id', 'season']

# Count matches played at home
home_matches = matches_df.groupby(['season', 'home_club_id']).size().reset_index(name='home_matches')

# Count matches played away
away_matches = matches_df.groupby(['season', 'away_club_id']).size().reset_index(name='away_matches')

# Rename columns for merging
home_matches.rename(columns={'home_club_id': 'club_id'}, inplace=True)
away_matches.rename(columns={'away_club_id': 'club_id'}, inplace=True)

# Merge the home and away match counts
total_matches = pd.merge(home_matches, away_matches, on=['season', 'club_id'], how='outer')

# Fill missing values with 0 (in case a team played only home or away matches)
total_matches.fillna(0, inplace=True)

# Calculate total matches played by each team per season
total_matches['MP'] = total_matches['home_matches'] + total_matches['away_matches']

# Now total_matches contains the total matches ('MP') played by each team per season

team_info = total_matches[total_matches['club_id'] == 31]
team_info

# Calculate the average matches played by each club over all seasons
average_matches_per_club = total_matches.groupby('club_id')['MP'].mean().reset_index(name='club_average_matches').round({'club_average_matches': 0})

# Display the resulting DataFrame
print(average_matches_per_club)
```
